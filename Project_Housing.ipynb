```python
from sklearn.base import BaseEstimator, TransformerMixin
```

```python
rooms_ix,bedroom_ix,populatioon_ix,household_ix = 3,4,5,6

class CombinedAttributeAdder(BaseEstimator,TransformerMixin):
    def __init__(self, add_bedroom_per_room=True):
        self.add_bedroom_per_room=add_bedroom_per_room

    def fit(self, X,y=None):
        return self

    def transform(self, X, y=None):

        rooms_per_househould = X[:,rooms_ix]/X[:,household_ix]
        population_per_household = X[:,populatioon_ix]/X[:,household_ix]

        if self.add_bedroom_per_room:
            bedrooms_per_room = X[:,bedroom_ix]/X[:,rooms_ix]
            return np.c_[X,rooms_per_househould,population_per_household,bedrooms_per_room]
        else:
             return np.c_[X,rooms_per_househould,population_per_household]



attr_adder =CombinedAttributeAdder(add_bedroom_per_room=False)

housing_extra_attribs = attr_adder.transform(housing.values)
housing_extra_attribs
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
```

```python
num_pipline = Pipeline([("imputer",SimpleImputer(strategy="median")),
                       ("attribs_adder", CombinedAttributeAdder()),
                       ("std_scaler",StandardScaler()),
                        ])
num_pipline.fit_transform(housing_number).shape
```

```python
from sklearn.compose import ColumnTransformer
```

```python
num_attribs = list(housing_number)
cat_attribs = ["ocean_proximity"]

full_pipeline = ColumnTransformer([("num", num_pipline , num_attribs),
                                   ("cat", OneHotEncoder(), cat_attribs)])


housing_prepared = full_pipeline.fit_transform(housing)
housing.head()
```
